```julia
function saveChain(name::AbstractString,P::AbstractChain)::nothing
    arr = toArray(P)
    open(name,"w+") do io
        DelimitedFiles.writedlm(io,arr,',')
    end
end
```

```julia
function saveTrajectory(name,P::PolygonalNew,angles,diheds)
    ns = length(P)+1
    nzeros = Int(ceil(log10(ns+1)))
    pnames = [lpad(i,nzeros,'0') for i in 1:ns]
    pnames = [string("p",i) for i in pnames]
    firstrow = join(pnames,',')
    firstrow = string(firstrow,"\n")
    newP = copy(P)
    open(name,"w+") do io
        write(io,firstrow)
        arr = toArray(newP)
        coordinates = [string(x) for x in reshape(transpose(arr),length(arr))]
        coordinates = join(coordinates,",")
        coordinates = string(coordinates,"\n")
        write(io,coordinates)
        for i in 1:length(angles)
            if diheds[i]!= 0 
                dihedralRotate!(newP,diheds[i],angles[i])
                arr = toArray(newP)
                coordinates = [string(x) for x in reshape(transpose(arr),length(arr))]
                coordinates = join(coordinates,",")
                coordinates = string(coordinates,"\n")
                write(io,coordinates)
            end
        end
    end
end
```
